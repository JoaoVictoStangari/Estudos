## **Aula 2**

```python
# criar uma variável só com as datas
datas = dado_transposto[:,0]
datas
```

```python
# criar uma variável apenas com os preços
precos = dado_transposto[:, 1:6]
precos
```

```python
# plotar um gráfico dos preços da cidade de Moskow de acordo a data
import matplotlib.pyplot as plt

plt.plot(datas, precos[:, 0])
```

```python
#alteração da variável data para melhor representação no gráfico
datas = np.arange(1, 88)

plt.plot(datas, precos[:, 0])
```

```python
# filtrar cada dado de preco por variáveis que representam a cidade
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
```

```python
# separação dos preços por ano da cidade de Moscown
Moscow_ano1 = Moscow[0:12]
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]
```

```python
# plotagem dos 4 anos separadamente da cidade de Moscow e adicionar legenda
plt.plot(np.arange(1,13,1), Moscow_ano1)
plt.plot(np.arange(1,13,1), Moscow_ano2)
plt.plot(np.arange(1,13,1), Moscow_ano3)
plt.plot(np.arange(1,13,1), Moscow_ano4)
plt.legend(["ano1", "ano2", "ano3", "ano4"])
```

```python
# verificar se dois conjuntos de dados são iguáis
np.array_equal(Moscow_ano3, Moscow_ano4)
```

```python
# verificar se um conjunto de dados esta dentro de um região expecificada
# o numero no final estipula o tamalho da região
np.allclose(Moscow_ano3, Moscow_ano4, 1)
```

```python
# quebra no gráfico em relação os valores
plt.plot(datas, Kaliningrad)
```
